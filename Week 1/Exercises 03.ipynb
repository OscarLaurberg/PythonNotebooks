```python
import matplotlib.pyplot as plt

student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}


plt.figure()
x_values = student_attendance.keys()
y_values = student_attendance.values()


plt.xlabel("Day", fontsize=14)
plt.ylabel("Attendance", fontsize=14)
plt.plot(list(x_values), list(y_values), linewidth=5)




```
